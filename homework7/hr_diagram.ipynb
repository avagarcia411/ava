```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
```

```python
# -- 3.2 Analyzing Imported Data -- 
# 1)
print(np.shape(data))

# 3)
magnitude = data[:, 0]
temperature = data[:, 1]

# 4)
print(magnitude.min()) # corresponds to relatively bright and easily visible stars like Polaris
print(magnitude.max()) # corresponds to stars that are too faint to be seen with the naked eye

# 5) 
print(temperature.min()) # corresponds to M-class stars
print(temperature.max()) # corresponds to A-class stars

# 6)
# The star with the magnitude of 2.24 is the brightest, and that star with the magnitude of 12.49 is the dimmest
```

```python
# -- 3.3 Plot the HR Diagram -- 
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

plt.figure(figsize = (8, 6))

plt.scatter(log_temperature, log_magnitude, label = "Star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

brightest_star = np.argmin(magnitude)
dimmest_star = np.argmax(magnitude)
hottest_star = np.argmax(temperature)
coldest_star = np.argmin(temperature)

plt.scatter(log_temperature[brightest_star], log_magnitude[brightest_star], color = 'black', marker = '*', s = 300, label = 'Brightest Star')
plt.scatter(log_temperature[dimmest_star], log_magnitude[dimmest_star], color = 'purple', marker = '*', s = 300, label = 'Dimmest Star')
plt.scatter(log_temperature[hottest_star], log_magnitude[hottest_star], color = 'green', marker = '*', s = 300, label = 'Hottest Star')
plt.scatter(log_temperature[coldest_star], log_magnitude[coldest_star], color = 'orange', marker = '*', s = 300, label = 'Coldest Star')

plt.legend()
plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")

plt.show
```
