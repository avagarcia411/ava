```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# -- 2: Matplotlib -- 
# 1) 
def make_sine_wave(x, A, w):
    y = A * np.sin(w * x)
    return y

# 2) 
x = np.linspace(0, 2*np.pi, 1000) # An array that goes from 0 to 2pi, with 1000 points

# 3)
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]

# 4)
fig, axes = plt.subplots(1, 3, figsize=(25, 5))

# 5, 6, and 7)
for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, 1)
    axes[0].plot(x, y, label=f"A={A}")
axes[0].set_title("Varying Amplitude (=1)", fontsize=16)
axes[0].set_xlabel("x", fontsize=14)
axes[0].set_ylabel("y", fontsize=14)
axes[0].legend(fontsize=12)
axes[0].grid(True)

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, 1, w)
    axes[1].plot(x, y, label=f"={w}")
axes[1].set_title("Varying Frequency (A=1)", fontsize=16)
axes[1].set_xlabel("x", fontsize=14)
axes[1].set_ylabel("y", fontsize=14)
axes[1].legend(fontsize=12)
axes[1].grid(True)

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w)
    axes[2].plot(x, y, label=f"A={A}, ={w}")
axes[2].set_title("Varying Amplitude and Frequency", fontsize=16)
axes[2].set_xlabel("x", fontsize=14)
axes[2].set_ylabel("y", fontsize=14)
axes[2].legend(fontsize=12)
axes[2].grid(True)

# 8)
plt.tight_layout()
plt.show()
```
