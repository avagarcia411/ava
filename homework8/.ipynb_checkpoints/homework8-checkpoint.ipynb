```python
import pandas as pd
```

```python
# -- 2: Pandas DataFrame -- 
df = pd.read_csv("../../course_assignments/homework8/state_data.csv")

# 4)
print(df.shape) # There are 52 rows, and 4 columns

# 5)
print(df.head(5)) 
print(df.tail(5))
# The different columns represent state, latitude, longitude, and name
# The first state in the dataset is Alaska
# The last state in the dataset is Wyoming
# There are 51 entries total
# There is more then 50 states total because DC is included

# 6)
median_latitude = df["latitude"].median()
print(median_latitude) # The median latitude is 39.4355155 degrees

median_longitude = df["longitude"].median()
print(median_longitude) # The median longitude is -89.093198 degrees

# 7)
df["lat_distance"] = (df["latitude"] - median_latitude).abs()

# 8)
df["long_distance"] = (df["longitude"] - median_longitude).abs()

# 9)
print(df.shape) # There are 52 rows and 6 columns

# 10)
closest_lat_state = df.loc[df["lat_distance"].idxmin()]
farthest_lat_state = df.loc[df["lat_distance"].idxmax()]
print(closest_lat_state) # The closest state to the median latitude is CO by 0.11
print(farthest_lat_state) # The farthest state to the median latitude is AK by 24.15

closest_long_state = df.loc[df["long_distance"].idxmin()]
farthest_long_state = df.loc[df["long_distance"].idxmax()]
print(closest_long_state) # The state closest to the median longitude is IL by 0.31
print(farthest_long_state) # The state farthest from the median longitude is HI by 66.57

# 11)
df.describe()
# I found it in the "Summarizing data: describe" section of the website. 
# This function computes a variety of summary statistics about a Series of columns in a DataFrame.
# It would be useful when you want to know information on all of your columns at the same time.
```

```python
# -- 3: Seaborn --
import matplotlib.pyplot as plt
import seaborn as sns

planets = sns.load_dataset('planets')
df = pd.DataFrame(planets)

# 5)
df.dropna(axis=1, inplace=True)

# 6)
print(df.shape) # The shape is (1035, 3)
# There are 1035 rows and 3 columns

print(df.head(5)) 
print(df.tail(5))
# The first columns name is "method" and it is the method that the planet was discovered
# the second column name is "number" and it is the number of planets in the system
# The third column name is "year" and it is the year the planet was discovered

# 7)
fig, axes = plt.subplots(2, 2, figsize=(10,8))

sns.scatterplot(x=df["method"], y=df["number"], ax=axes[0,0], color="red")
axes[0,0].set_title("Number of Planets in System vs. Method of Discovery")
axes[0,0].set_xlabel("Method of Discovery")
axes[0,0].set_ylabel("Number of Planets in System")
# axes[0,0].legend() I was getting an error trying to put a legend in so I could not figure it out :(

sns.scatterplot(x=df["number"], y=df["method"], ax=axes[0,1], color="green")
axes[0,1].set_title("Method of Discovery vs. Number of Planets in System")
axes[0,1].set_xlabel("Number of Planets in System")
axes[0,1].set_ylabel("Method of Discovery")

sns.histplot(df["year"], ax=axes[1,0], color="blue")
axes[1,0].set_title("Year of Planet Discovery")
axes[1,0].set_xlabel("Year of Planet Discovery")
axes[1,0].set_ylabel("Counts")

sns.boxplot(x=df["number"], ax=axes[1,1], color="purple")
axes[1,1].set_title("Number of Planets in System")
axes[1,1].set_xlabel("Number of Planets in System")

plt.tight_layout()
plt.show()

# 9)
# There were only 3 columns to choose from so I did not have much choice in which ones I chose for what. but I chose the first two columns for the first two subplots because I thought they were the most interesting.
# I learned that there is not really any correlation between the number of planets in a system and the method of discovery, there were more planets that were discovered recently compared to a long time ago due to imporovements from technology, and there are 4 outliers for the number of planets in a system.
```
